### 변수명 바꾸기

```python
# 기존데이터를 수정할땐 copy 만들기 추천
mpg_copy = mpg_data.copy()
```

```python
mpg_copy
```

```python
# .rename()
mpg_copy = mpg_copy.rename(columns= {"manufacturer" : "제조사", "cyl": "실린더" , "cty" :"도시" , "hwy" : "고속도로" , "fl" : "연료"})
mpg_copy
```

```python
# 도시와 고속도로의 연비를 구한 후에 평균
# 데이터 접근
# 1. 도시 접근
display(mpg_copy['도시'])

# 2. 고속도로 연비 접근
display(mpg_copy['고속도로'])

# 도시 + 고속도로 / 2
display((mpg_copy['도시'] + mpg_copy['고속도로']) /2 )
```

```python
# loc, iloc 인덱서 사용
mpg_copy.loc[0]
```

```python
mpg_copy.iloc[0]
```

```python
mpg_index = mpg_copy.set_index('제조사')
mpg_index
```

```python
mpg_index.iloc[0]
```

```python
mpg_index.loc['audi']
```

```python
# loc - 슬라이싱
mpg_index.loc['audi' : 'hyundai']
```

```python
mpg_index.loc['audi' : 'hyundai', '도시' : '고속도로']
```

```python
mpg_index.iloc[0 : 10]
```

### 불리언 인덱싱

```python
# 제조사만 보고 싶은경우
mpg_copy[mpg_copy['도시'] > 20]['제조사']
```

```python
# query 함수
mpg_copy.query('도시 > 20')
```

### 파생변수 만들기

- 컬럼 추가
- 데이터명['새로운 컬럼명'] = 데이터 대입

```python
mpg_mean = (mpg_copy['도시'] + mpg_copy['고속도로']) / 2
mpg_mean
```

```python
mpg_copy['평균'] = mpg_mean
```

```python
mpg_copy
```

```python
mpg_copy['평균']
```

```python
mpg_copy['평균'].mean()
```

```python
# 4. exam 데이터의 각 평균값 구하기
exam_data_mean = round(exam_data.loc[:,'math':'science'].mean(axis='columns'),1)
exam_data_mean
```

```python
# 5. 평균 컬럼 파생변수 만들기
exam_data['평균'] = exam_data_mean
```

```python
exam_data
```

```python
import numpy as np
```

```python
# 평균 점수가 60점 이상인 사람 pass, fail
# pass, fail의 값도 확인
exam_data['test 결과'] = np.where(exam_data['평균']>=60, 'pass', 'fail')
exam_data
```

```python
exam_data['test 결과'].value_counts()
```

```python
# 평균 점수가 80이상 1등급 70 2등급 60 3등급 재시험
```

```python
def makeGrade(df): 
    grade = df['평균']
    
    if grade >= 80:
        return "1등급"
    elif grade >= 70:
        return "2등급"
    elif grade >= 60:
        return "3등급"
    else:
        return "재시험"
```

```python
exam_data["grade"] = exam_data.apply(makeGrade, axis=1)
exam_data
```

```python
exam_data['grade'].value_counts().sort_index()
```

```python
exam_data['grade'].value_counts().sort_index()
```

```python
# nclass를 그룹으로 묶고 그룹별 math의 평균값을 구하기
exam_data.groupby('nclass').agg(math_mean = ('math','mean'))
```

```python
exam_data.groupby('nclass').mean(True)
```

```python
# nclass를 그룹으로 묶고 그룹별 result pass/fail인 사람들의 수를 구하기
exam_data.groupby(['nclass', 'test 결과']).agg(r = ('test 결과','count'))
```

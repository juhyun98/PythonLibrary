# Line Plot

```python
# line plot을 그리고 위해서는 x축 y축이 필요
x = np.arange(1,6) # x = [1,2,3,4,5]
np.random.seed(3)
y = np.random.randint(1,10, size = 5) # y = [?,?,?,?,?]

print(x)
print(y)
```

```python
# plt.plot(x축, y축)

plt.plot(x, y)
```

```python
# plt.plot(y축)
# -- x축은 자동으로 0~ y축의 개수 만큼 증가
plt.plot(y)
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y, ls = ':')
```

```python
plt.plot(x, y, ls =':', c = 'r', marker = 'D', mec = 'b', mfc = 'b')

# 만약 그래프가 안나오면 plt.show() 사용해주기
# 쥬피터 노트북은 출력을 지원하는 개발환경이라서 사용 안해줘도 출력
```

```python
# 한 화면에 여러개의 그래프를 출력

# 방법 : plt.subplot(row, col, index)

plt.subplot(2,2,1)
plt.plot(x, y, ls = '-', c = 'b', marker ='D', ms = 10, mec = 'c', mew = 4, mfc = 'y')

plt.subplot(2,2,2)
plt.plot(x, y, ls = '--', c = 'm', lw = 5)

plt.subplot(2,2,3)
plt.plot(x,y, ls = '-.')

plt.subplot(2,2,4)
plt.plot(x,y, ls = ':')
```

```python
# 여러 그래프를 한 창에 그려지게(그래프 겹치기)

one_x = np.arange(1,4) # [1,2,3]
one_y = np.array([1,2,3])
one_y2 = np.array([4,5,1])
one_y3 = np.array([3,6,2])

plt.plot(one_x, one_y)
plt.plot(one_x, one_y2)
plt.plot(one_x, one_y3)
```

```python
# plt.show()

plt.plot(one_x, one_y2)
plt.plot(one_x, one_y3)
```

# legend()

```python
test_x = np.arange(1,6)
np.random.seed(5)
test_y = np.random.randint(1,10, size = 5)

test_x1 = np.arange(1,6)
np.random.seed(10)
test_y1 = np.random.randint(1,10, size = 5)

print(test_x)
print(test_y)
print(test_x1)
print(test_y1)
```

```python
# legend() 사용방법
# plt.legend()
# 단, plt.plot을 그릴때 해당 plot의 label값을 지정 해주어야합니다!
#      ㄴ plt.plot(x,y, label = '라벨값 지정')

plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x1, test_y1, label = 'y1')
plt.legend(loc = 'upper right')
```

# xlabel(), ylabel()

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# plt.xlabel('이름')
# plt.ylabel('이름')


plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x1, test_y1, label = 'y1')
plt.legend(loc = 'upper right')

plt.xlabel('x축')
plt.ylabel('y축')
```

# xlim(), ylim()

```python
# xlim, ylim(시작, 끝)

plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x1, test_y1, label = 'y1')
plt.legend(loc = 'upper right')

plt.xlabel('x축')
plt.ylabel('y축')

plt.xlim(0, 6)
plt.ylim(0, 10)
```

# xticks, yticks

```python
# 사용법 xticks, yticks(list, 배열)


plt.plot(test_x, test_y, label = 'y')
plt.plot(test_x1, test_y1, label = 'y1')
plt.legend(loc = 'upper right')

plt.xlabel('x축')
plt.ylabel('y축')

plt.xticks(range(1,6)) # 1,2,3,4,5
plt.yticks(range(1,11,2)) # 1,3,5,7,9
```

### 컬럼 -- grade (등급)
- 평균값이 25이상이면 A등급
- 20이상 B등급
- 그 외 C등급

```python
mpg_grade = np.where(mpg_copy['평균'] >= 25, 'A등급', 
                     np.where(mpg_copy['평균']>=20, 'B등급','C등급'))
```

```python
mpg_copy['grade'] = mpg_grade
mpg_copy
```

```python
mpg_copy['grade'].value_counts()
```

## apply()라는 함수
- np.where의 조건식 복잡해지는 것을 간단하게 작성 할 수 있도록 하는 함수

```python
# apply() 사용법

# 매개변수가 함수
# 함수를 만들어서 함수명만 작성 해주시면 됩니다.
def makeGrade(df): # 함수 실행시 조건이 될 시리즈가 필요! 매개변수로 df 작성
    grade = df['평균']
    
    if grade >= 25:
        return "A등급"
    elif grade >= 20:
        return "B등급"
    elif grade >= 15:
        return "C등급"
    else:
        return "F등급"
```

- axis = 0(행) or 1(열, 축)

```python
# apply()
mpg_copy["grade2"] = mpg_copy.apply(makeGrade, axis=1)
mpg_copy
```

```python
mpg_copy['grade2'].value_counts().sort_index()
```

```python
# pd.merge()

# db - join inner, outer - left right

# how
# left = df1이 기준
# right = df2가 기준
# outer - 데이터를 모두 합치는데 만약에 df1이나 df2와 일치하지 않는 데이터가 존재 시
#       ㄴ nan값이 들어가게 됨
# test3로 확인

# inner - 교집합 부분만 표시 (일치하지 않는 데이터는 merge X)

# on - 기준이 될 컬럼
# pd.merge(df1, df2, how = '')
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5,6],'midterm':[60,80,70,90,85,80]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'final':[70,83,65,95,80]})
```

```python
test3 = pd.merge(test1, test2, how = 'outer', on='id')
test3
```

```python
test4 = pd.merge(test1, test2, how = 'inner', on='id')
test4
```
